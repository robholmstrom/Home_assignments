```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate, GridSearchCV
from sklearn.metrics import roc_auc_score
import joblib
from dask.distributed import Client, progress
from dask_ml.model_selection import train_test_split
import dask.dataframe as dd
import pandas as pd
import warnings

import dask.array as da
```

```python
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y =True)
```

```python
Xda = da.array(X)
yda = da.array(y)
```

```python
Xdf = dd.DataFrame(Xdf)
```

```python
Xdf
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xda, yda, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
with joblib.parallel_backend('dask'):
    lr_clf = LogisticRegression(solver='lbfgs', multi_class = 'ovr', max_iter = 1000)
    lr_cv_scores = cross_val_score(lr_clf, X_train.compute(), y_train.compute(), cv = 5)

print('Accuracy scores for the 5 folds: ', lr_cv_scores)
print('Mean cross validation score: {:.3f}'.format(lr_cv_scores.mean()))
```

```python
from sklearn.ensemble import RandomForestClassifier
with joblib.parallel_backend('dask'):    
    rf_clf = RandomForestClassifier(n_estimators=24)
    rf_cv_scores = cross_val_score(rf_clf, X_train.compute(), y_train.compute() , cv = 5 )

print('Accuracy scores for the 5 folds: ', rf_cv_scores)
print('Mean cross validation score: {:.3f}'.format(rf_cv_scores.mean()))
```
