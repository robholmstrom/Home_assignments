```
# 1. Merge these two DataFrames into a single DataFrame called 'census'
import pandas as pd

# Create separate tables:
populations = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
print(populations.head(10))
states = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')
print(states.head(10))
# Merge tables:
census = pd.merge(populations, states)

census.head(10)


```

```
# 2. Re-shape 'census' so that one column contains 
# all population measures, and another the year attributes

census_melt = pd.melt(frame = census, id_vars = ['state'], var_name = 'year', value_name = 'population')
census_melt.tail(10)
```

```
melted_df = pd.melt(census, id_vars=['state'], 
        var_name='year',
        value_name='population',
        value_vars=['2010','2011','2012','2013','2014','2015','2016'])
```

```
 # 3. Group the data by year and summarize it
 melted_df.groupby('year').mean()
```
